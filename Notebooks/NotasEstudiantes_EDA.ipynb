Ambos DataSets contienen la misma cantidad de filas, la diferencia son las columnas extras que contiene el DataSet 2 (DF_Dos). Se resalta que la columna "Unnamed: 0" contiene valores duplicados ya que el indice se renicia después de 999. Se uniran ambos datasets para evitar el problema de los valores duplicados se creará una columna en cada DatSet con un valor único y esa columna servirá para realizar el merge.

```python
# Crear columna índice con valores únicos para DataSet Uno
DF_Uno["ID"] = range(1, len(DF_Uno)+1)

# Colocar la columna ID al inicio
columna = ["ID"] + [col for col in DF_Uno.columns if col != "ID"]
DF_Uno = DF_Uno[columna]
```

```python
DF_Uno.head(10)
```

```python
# Crear columna índice con valores únicos para DataSet Dos
DF_Dos["ID"] = range(1, len(DF_Dos)+1)

# Colocar la columna ID al inicio
columna = ["ID"] + [col for col in DF_Dos.columns if col != "ID"]
DF_Dos = DF_Dos[columna]
```

```python
DF_Dos.head(10)
```

Las columnas a considerar del segundo DataFrame son:
- Medio De Transporte
- Estado Civil
- Practia De Deporte
- Horas De Estudio Semanalmente

```python
NotasEstudiantes = pd.merge(
    DF_Uno,
    DF_Dos[["ID", "ParentMaritalStatus", "PracticeSport", "TransportMeans", "WklyStudyHours"]],
    on="ID",
    how="left"
)

NotasEstudiantes
```

Se procederá a eliminar la columna "Unnamed: 0" y a cambiar los nombres de las columnas a español.

```python
# Eliminar columna 'Unnamed: 0'
del NotasEstudiantes["Unnamed: 0"]
```

```python
# Cambiar nombre de columnas a español
NotasEstudiantesDos = NotasEstudiantes.rename(columns=
                        {"Gender":"Género",
                         "EthnicGroup":"Grupo Étnico",
                         "ParentEduc":"Nivel Educativo Padres",
                         "LunchType":"Tipo De Almuerzo",
                         "TestPrep":"Preparación Para Examen",
                         "MathScore":"Nota Matemáticas",
                         "ReadingScore":"Nota Lectura",
                         "WritingScore":"Nota Escritura",
                         "ParentMaritalStatus":"Estado Civil Padres",
                         "PracticeSport":"Práctica Deportiva",
                         "TransportMeans":"Medio De Transporte",
                         "WklyStudyHours":"Horas De Estudio Semanal"})

NotasEstudiantesDos.head(3)
```

```python
NotasEstudiantesDos[["Estado Civil Padres","Práctica Deportiva","Medio De Transporte", "Horas De Estudio Semanal"]].isnull().sum()
```

Estas son las columnas que contienen valores nulos, esto se modificará para que no afecte en las visualizaciones:
 - Para la columna "Estado Civil Padres" los valores nulos se modificaran a not specified
 - Para la columna "Práctica Deportiva" los valores nulos se modificaran a sometimes debido a la alta cantidad de registros con estos datos
 - Para la columna "Medio De Transporte" los valores nulos se modificaran a walking 
 - Para la columna "Horas De Estudio Semanal" los valores nulos se modificaran a 5 - 10 debido a la alta cantidad de registros con estos datos

```python
# Cambio de valores nulos
# 1. Columna Estado Civil Padres
NotasEstudiantesDos["Estado Civil Padres"] = (
    NotasEstudiantesDos["Estado Civil Padres"].fillna("not specified")
)

# 2. Columna Práctiva Deportiva
modaDeporte = NotasEstudiantesDos["Práctica Deportiva"].mode()[0]

NotasEstudiantesDos["Práctica Deportiva"] = (
    NotasEstudiantesDos["Práctica Deportiva"].fillna(modaDeporte)
)

# 3. Columna Medio De Transporte
NotasEstudiantesDos["Medio De Transporte"] = (
    NotasEstudiantesDos["Medio De Transporte"].fillna("walking")
)

# 4. Columna Horas De Estudio Semanal
modaEstudioSemanal = NotasEstudiantesDos["Horas De Estudio Semanal"].mode()[0]

NotasEstudiantesDos["Horas De Estudio Semanal"] = (
    NotasEstudiantesDos["Horas De Estudio Semanal"].fillna(modaEstudioSemanal)
)
```

```python
NotasEstudiantesDos.isnull().sum()
```

El DataFrame ya no contiene valores nulos

# Datos Cálculados

```python
Grupos = NotasEstudiantesDos.groupby("Grupo Étnico")["Grupo Étnico"].count()
print(f"Cantidad de alumnos pertenecientes a cada grupo\n{Grupos}")
```

```python
GruposMatematicaPromedio = NotasEstudiantesDos.groupby("Grupo Étnico")["Nota Matemáticas"].mean()
print(f"Promedio de notas por cada grupo para el examen de Matemáticas \n{round(GruposMatematicaPromedio, 1)}")
```

```python
GruposLecturaPromedio = NotasEstudiantesDos.groupby("Grupo Étnico")["Nota Lectura"].mean()
print(f"Promedio de notas por cada grupo para el examen de Lectura \n{round(GruposLecturaPromedio, 1)}")
```

```python
GruposEscrituraPromedio = NotasEstudiantesDos.groupby("Grupo Étnico")["Nota Escritura"].mean()
print(f"Promedio de notas por cada grupo para el examen de Escritura \n{round(GruposEscrituraPromedio, 1)}")
```

Se observa que el grupo A es el que tiene los promedios más bajos en los 3 examenes, esto se puede deber a varios factores:
- Tipo de transporte en el que se moviliza
- Horas de estudio a la semana invertidas
