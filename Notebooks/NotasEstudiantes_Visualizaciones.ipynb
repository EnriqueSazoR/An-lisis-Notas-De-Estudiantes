```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

# Importar DataSet con datos Limpios

```python
Ruta = r"C:\Users\Luis Sazo\Desktop\Análisis Notas Estudiantes\Data\NotasEstudiantesLimpio.csv"
NotasEstudiantesDF = pd.read_csv(Ruta)
NotasEstudiantesDF.head(10)
```

```python
NotasEstudiantesDF["Grupo Étnico"].unique()
```

# Visualizaciones Notas

1. Matemáticas

```python
def GraficaMatematicas(df, columna):
    fig, ax = plt.subplots(1, 5, figsize=(17, 3))
    grupos = ["group A", "group B", "group C", "group D", "group E"]

    # Gráficar en cada subplot
    for i, g in enumerate(grupos):
        df_g = NotasEstudiantesDF[NotasEstudiantesDF["Grupo Étnico"] == g]
        
        notas = df_g["Nota Matemáticas"]
        media = notas.mean()
        
        ax[i].scatter(notas, [1]*len(notas), s=10, alpha=0.25, color="green")
        ax[i].axvline(media, color="black", linestyle="--", linewidth=2)
        
        antes = (notas <= media).sum()
        despues = (notas >= media).sum()
        total = len(notas)
        
        avgAntes = antes / total * 100
        avgDespues = despues / total * 100
        
        ax[i].text(media - 5, 1.02, f"{round(avgAntes)}%", ha="right", fontsize=10)
        ax[i].text(media + 5, 1.02, f"{round(avgDespues)}%", ha="left", fontsize=10)
        
        ax[i].set_title(f"{g}")
        ax[i].set_xlabel("Matemáticas")
        ax[i].set_yticks([])

    plt.tight_layout()
    
    # Guardar Gráfico
    plt.savefig("../Gráficas/DispersiónNotasMatemáticas.png", dpi=300)
    plt.show();


GraficaMatematicas(NotasEstudiantesDF, NotasEstudiantesDF["Nota Matemáticas"])
```
